### Preguntas:

1. ¿Existen columnas que podrían ser categóricas pero aparecen como tipo `object`?

2. ¿Hay valores que parecen inconsistentes o fuera de rango?  
    - No pude identificar valores atípicos o fuera de rango. Utilicé la función `Zscore_outlier`, que trabajamos con el Dr. Tomás, para detectar datos atípicos. Además, usé `data.describe()` para hacer una inspección visual, revisando los valores mínimos y máximos en busca de anomalías.  

```python
import numpy as np

out = []
def Zscore_outlier(df):
  m = np.mean(df)
  sd = np.std(df)
  for i in df:
    z = (i - m)/sd
    if np.abs(z) > 3:
      out.append(i)
  print("Outliers:",out)
```

```python
Zscore_outlier(data['Age'])
Zscore_outlier(data['Social_Support_Score'])
Zscore_outlier(data['Anxiety_Score'])
Zscore_outlier(data['Depression_Score'])
Zscore_outlier(data['Stress_Level'])
Zscore_outlier(data['Family_History_Mental_Illness'])
Zscore_outlier(data['Chronic_Illnesses'])
Zscore_outlier(data['Therapy'])
Zscore_outlier(data['Meditation'])
Zscore_outlier(data['Financial_Stress'])
Zscore_outlier(data['Work_Stress'])
Zscore_outlier(data['Self_Esteem_Score'])
Zscore_outlier(data['Life_Satisfaction_Score'])
Zscore_outlier(data['Loneliness_Score'])
data.describe()
```

## Paso 3: Limpieza de Datos

### Preguntas:

1. ¿Qué porcentaje del dataset contiene valores nulos?
    - 6.27% del dataset presenta valores nulos
2. ¿Qué estrategia decidiste aplicar para tratar los valores nulos y por qué?
    - Decidí reemplazar los valores `NaN` por la categoría `'Unknown'` en las columnas `Medication_Use` y `Substance_Use`. Esto para poder distinguir entre los datos faltantes y las categorías válidas. Luego, utilicé `LabelEncoder` para convertir estas categorías a valores numéricos, lo que ayudaria en el análisis y evita que los valores nulos se eliminen y poderlos distinguir entre las categorías válidas. me parecio más adecuado que intentar imputar los valores, ya que los datos faltantes representan mas de 15% 





```python
#Columnas que tienen valores nulos
porcentaje_nulos = (data.isnull().sum() / len(data)) * 100
porcentaje_nulos
```

```python
#% del dataset que tiene valores nulos
porcentaje_nulos_total = (data.isnull().sum().sum() / data.size) * 100  
porcentaje_nulos_total
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()

dataCopy = data;

# Reemplazar NaN con una categoria 'Unknown' para 'Medication_Use' y 'Substance_Use'
dataCopy['Medication_Use'] = dataCopy['Medication_Use'].fillna('Unknown')
dataCopy['Substance_Use'] = dataCopy['Substance_Use'].fillna('Unknown')

dataCopy['Medication_Use_Num'] = encoder.fit_transform(dataCopy['Medication_Use'])
dataCopy['Substance_Use'] = encoder.fit_transform(dataCopy['Substance_Use'])
dataCopy
```

## Paso 4: Análisis Estadístico
En este paso, vas a calcular algunas estadísticas básicas para tres columnas importantes: `Anxiety_Score`, `Depression_Score` y `Stress_Level`. Queremos saber cuál es la media, la mediana y la desviación estándar de estos valores. Esto te permitirá tener una idea general sobre cómo están distribuidos estos puntajes.

### Preguntas:

1. ¿Cuál de los tres puntajes tiene mayor variabilidad?
    - El puntaje con mayor variabilidad es `Anxiety_Score`, ya que tiene la mayor desviación estándar de `5.91`. Esto indica que los puntajes de depresión están más dispersos en comparación con las otras variables. 
2. ¿Qué observas sobre la diferencia entre media y mediana en cada caso? ¿Qué te dice esto sobre la distribución?
    - `Anxiety_Score` y `Stress_Level` tienen una media y mediana casi iguales, lo que indica que la distribución es **simétrica**. En cambio, para `Depression_Score`, la media es un poco menor que la mediana, lo que sugiere que los datos están ligeramente sesgados 









```python
columns = ['Anxiety_Score', 'Depression_Score', 'Stress_Level']

mean_values = dataCopy[columns].mean()
median_values = dataCopy[columns].median()
std_values = dataCopy[columns].std()

print("Media:\n", mean_values)
print("\nMediana:\n", median_values)
print("\nDesviación estándar:\n", std_values)
```

Luego, clasifica a las personas según las horas de sueño que reportan (menos de 5 horas, entre 5 y 7 horas, y más de 7 horas) y analiza cómo varía el nivel de estrés en cada grupo.

### Preguntas:

1. ¿Qué porcentaje del total presenta niveles altos de ansiedad?

```python
con_alta_ansiedad = dataCopy[dataCopy['Anxiety_Score'] > 19]
con_alta_ansiedad.head()
```

```python

con_alta_ansiedad = con_alta_ansiedad.copy()  
con_alta_ansiedad.loc[con_alta_ansiedad['Sleep_Hours'] < 5, 'horas_dormidas'] = 'Menos de 5 horas'
con_alta_ansiedad.loc[(con_alta_ansiedad['Sleep_Hours'] >= 5) & (con_alta_ansiedad['Sleep_Hours'] <= 7), 'horas_dormidas'] = 'Entre 5 y 7 horas'
con_alta_ansiedad.loc[con_alta_ansiedad['Sleep_Hours'] > 7, 'horas_dormidas'] = 'Más de 7 horas'

stress_by_sleep = con_alta_ansiedad.groupby('horas_dormidas')['Stress_Level'].mean()
print(stress_by_sleep)
```
